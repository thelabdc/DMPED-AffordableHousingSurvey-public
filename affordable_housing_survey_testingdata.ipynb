# Affordable Housing Survey - Data Analysis

In this notebook I created charts and visuals using our sample data. There are five key features that we are particularly interested in analyzing: Ward `Ward`, Number of Moves `Moves#`, HouseholdIncome `HHInc`, Number of Bedrooms `Bedrooms#` and Number of Tenants `Tenants#`. 

## Import data

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('test.csv')
df=df[['WARD','NumberMoves',' HouseholdInc ','Bedrooms','Tenants']]
```

```python
df.columns=['Ward','Moves#','HHInc','Bedrooms#','Tenants#']
df = df.sort_values(['HHInc', 'Ward'], ascending=[1, 1])
df
```

## An Overview of the Sample Data

```python
sns.boxplot(data=df,orient='h')
```

## Topic 1. Relation between Income Level and Residency Mobility
Features used: `Moves#`, `HHInc`

```python
sns.barplot(x='HHInc',y='Moves#',data=df).set_title('Average Moves by Hosuehold Income Range')
```

```python
g=sns.FacetGrid(df,col='HHInc')
g=g.map(plt.hist,"Moves#")
plt.subplots_adjust(top=0.7)
g.fig.suptitle('Average Moves of Each Hosuehold Income Range')
```

## Topic 2. Relation between Ward and Residency Mobility
Features used: `Ward`, `Moves#`

```python
sns.barplot(x='Ward',y='Moves#',ci=None,data=df).set_title('Average Moves by Ward')
```

```python
g=sns.FacetGrid(df,col='Ward')
g=g.map(plt.hist,"Moves#")
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Average Moves of Each Ward')  
```

### ( Double Click to View Full Image)

```python
plt.subplots(figsize=(20,15))
sns.barplot(x='HHInc',y='Moves#',ci=None,hue='Ward',data=df).set_title('Average Moves by Income Level and Ward')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.set(font_scale=4)  
```

```python
plt.subplots(figsize=(20,15))
sns.barplot(x='Ward',y='Moves#',hue='HHInc',ci=None,data=df).set_title('Average Moves by Ward and Income Level')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.set(font_scale=3)
```

```python
# Reset Style
sns.reset_orig()
```

## Topic 3. Relation between Number of Bedrooms and Tenants
Features used: `Bedrooms#`, `Tenants#`

```python
sns.set_style("whitegrid")
h=sns.lmplot(x='Tenants#',y='Bedrooms#',hue='HHInc',data=df)
h.set(ylim=(0,4), yticks=[1,2,3,4])
```

```python
h=sns.swarmplot(x='Tenants#',y='Bedrooms#',data=df)
h.set(ylim=(0,4), yticks=[1,2,3,4])
```

```python
sns.set_style("whitegrid")
sns.factorplot(x='Tenants#',y='Bedrooms#',data=df)
```

```python
sns.violinplot(x='Tenants#',y='Bedrooms#',data=df)
```
